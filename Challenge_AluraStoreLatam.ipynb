### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_ingresos_tienda1 = tienda['Precio'].sum()
total_ingresos_tienda2 = tienda2['Precio'].sum()
total_ingresos_tienda3 = tienda3['Precio'].sum()
total_ingresos_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total de Tienda 1: {total_ingresos_tienda1}")
print(f"Ingreso total de Tienda 2: {total_ingresos_tienda2}")
print(f"Ingreso total de Tienda 3: {total_ingresos_tienda3}")
print(f"Ingreso total de Tienda 4: {total_ingresos_tienda4}")
```

# 2. Ventas por categoría

```python
# Ventas por categoría en Tienda 1
print("Ventas por categoría en Tienda 1:")
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
print(ventas_por_categoria_tienda1)
print("\n---\n")

# Ventas por categoría en Tienda 2
print("Ventas por categoría en Tienda 2:")
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
print(ventas_por_categoria_tienda2)
print("\n---\n")

# Ventas por categoría en Tienda 3
print("Ventas por categoría en Tienda 3:")
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
print(ventas_por_categoria_tienda3)
print("\n---\n")

# Ventas por categoría en Tienda 4
print("Ventas por categoría en Tienda 4:")
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
print(ventas_por_categoria_tienda4)
```

```python
import pandas as pd

# Combinar las ventas por categoría de todas las tiendas en un solo DataFrame para fácil comparación
ventas_categorias_comparativo = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
}).fillna(0).astype(int) # Rellenar NaN con 0 y convertir a entero

print("Comparativo de Ventas por Categoría entre Tiendas:")
display(ventas_categorias_comparativo)

# Opcional: Ordenar para ver las categorías más vendidas en general o por tienda
print("\n--- Las 5 Categorías más vendidas en total ---")
display(ventas_categorias_comparativo.sum(axis=1).nlargest(5))
```

# 4. Productos más y menos vendidos

```python
print("Productos más y menos vendidos por Tienda:")

todas_las_tiendas = {'Tienda 1': tienda, 'Tienda 2': tienda2, 'Tienda 3': tienda3, 'Tienda 4': tienda4}

for nombre_tienda, df_tienda in todas_las_tiendas.items():
    print(f"\n--- {nombre_tienda} ---")

    # Productos más vendidos
    productos_vendidos = df_tienda['Producto'].value_counts()
    if not productos_vendidos.empty:
        mas_vendido = productos_vendidos.index[0]
        cantidad_mas_vendido = productos_vendidos.iloc[0]
        print(f"Producto más vendido: '{mas_vendido}' ({cantidad_mas_vendido} unidades)")

        # Productos menos vendidos (si hay más de uno)
        if len(productos_vendidos) > 1:
            menos_vendido = productos_vendidos.index[-1]
            cantidad_menos_vendido = productos_vendidos.iloc[-1]
            print(f"Producto menos vendido: '{menos_vendido}' ({cantidad_menos_vendido} unidades)")
        else:
            print("No hay suficientes productos para determinar el menos vendido.")
    else:
        print("No hay datos de productos para esta tienda.")
```

```python
import pandas as pd

# Lista para almacenar los productos más y menos vendidos de cada tienda
comparativo_productos_vendidos = []

for nombre_tienda, df_tienda in todas_las_tiendas.items():
    productos_vendidos = df_tienda['Producto'].value_counts()

    mas_vendido = 'N/A'
    cantidad_mas_vendido = 0
    menos_vendido = 'N/A'
    cantidad_menos_vendido = 0

    if not productos_vendidos.empty:
        mas_vendido = productos_vendidos.index[0]
        cantidad_mas_vendido = productos_vendidos.iloc[0]

        if len(productos_vendidos) > 1:
            menos_vendido = productos_vendidos.index[-1]
            cantidad_menos_vendido = productos_vendidos.iloc[-1]

    comparativo_productos_vendidos.append({
        'Tienda': nombre_tienda,
        'Producto Más Vendido': mas_vendido,
        'Cantidad Más Vendido': cantidad_mas_vendido,
        'Producto Menos Vendido': menos_vendido,
        'Cantidad Menos Vendido': cantidad_menos_vendido
    })

# Crear un DataFrame a partir de la lista
df_comparativo = pd.DataFrame(comparativo_productos_vendidos)

print("Tabla Comparativa de Productos Más y Menos Vendidos por Tienda:")
display(df_comparativo)
```
